**Давайте добавим новый признак.**

```python
def age_category(age):
    '''
    < 30 -> 1
    >= 30, <55 -> 2
    >= 55 -> 3
    '''
    if age < 30:
        return 1
    elif age < 55:
        return 2
    elif age >= 55:
        return 3
```

```python
age_categories = [age_category(age) for age in data.Age]
data['Age_category'] = age_categories
```

**Другой способ сделать это с `apply`.**

```python
data['Age_category'] = data['Age'].apply(age_category)
```

**5. Это правда, что пассажиры моложе 30 лет. выжили чаще, чем те, кому больше 60 лет. Каковы доли выживших людей среди молодых и пожилых людей?**

**8. Как средний возраст мужчин / женщин зависит от `Pclass`? Выберите все правильные утверждения:**
- В среднем мужчины 1 класса старше 40 лет
- В среднем женщины 1 класса старше 40 лет
- Мужчины всех классов в среднем старше, чем женщины того же класса
- В среднем, пассажиры первого класса старше, чем пассажиры 2-го класса, которые старше, чем пассажиры 3-го класса.

```python
# You code here
people1Class = data[(data['Pclass'] == 1)]['Age'].mean()
people2Class = data[(data['Pclass'] == 2)]['Age'].mean()
people3Class = data[(data['Pclass'] == 3)]['Age'].mean()

men1Class = data[(data['Pclass'] == 1) & (data['Sex'] == 'male')]['Age'].mean()
men2Class = data[(data['Pclass'] == 2) & (data['Sex'] == 'male')]['Age'].mean()
men3Class = data[(data['Pclass'] == 3) & (data['Sex'] == 'male')]['Age'].mean()

print('Average Age Men of 1 Class', men1Class, 'The Answer is', 'Yes' if men1Class > 40 else 'No')
women1Class = data[(data['Pclass'] == 1) & (data['Sex'] == 'female')]['Age'].mean()
women2Class = data[(data['Pclass'] == 2) & (data['Sex'] == 'female')]['Age'].mean()
women3Class = data[(data['Pclass'] == 3) & (data['Sex'] == 'female')]['Age'].mean()

print('Average Age Women of 1 Class', women1Class, 'The Answer is', 'Yes' if women1Class > 40 else 'No')

print('Yes' if men1Class > women1Class and men2Class > women2Class and men3Class > women3Class else 'No')
print('Yes' if people1Class > people2Class and people2Class > people3Class else 'No')
```
